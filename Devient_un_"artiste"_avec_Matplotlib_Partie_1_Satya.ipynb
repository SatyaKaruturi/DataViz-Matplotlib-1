# Mission 2

Trace 2 graphiques et fait les appraître l'un en dessous de l'autre :

- Un carré : Les lignes qui le composent seront des pointillés de couleur verte.
- Un parralélograme : Les lignes qui le composent seront des traits épais de couleur rouge.

```
import numpy as np
import matplotlib.patches as patches

#Carré variables
x = [1,2,2,1,1]
y = [1,1,2,2,1]

#Parrelélogram varibles
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal')

#Carré
plt.plot(x, y, 'g:')
plt.xlim(-1, 3)
plt.ylim(-1, 3)

#Parrelélogram
x = [1,2.2,2.6,1.5]
y = [-0.5,-0.5,0.5,0.5]
ax.add_patch(patches.Polygon(xy=list(zip(x,y)), fill=False, edgecolor='r'))

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))               
nse2 = np.random.randn(len(t))               

s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('time')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('coherence')

cxy, f = axs[0].cohere(s1, s2, 256, 1. / dt)
axs[1].plot(t, s2)

axs[0].yaxis.set_label_position("right")
axs[1].yaxis.set_label_position("right")

fig.tight_layout()
plt.show()
```
